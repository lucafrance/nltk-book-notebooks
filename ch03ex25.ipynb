```python
import re

import nltk
```

```python
def word_to_pig_latin(word):
    consonants = "bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ"
    
    # Check if the word only includes consonants to avoid infnite loops
    all_consonants_word = True
    for c in word:
        if c not in consonants:
            all_consonants_word = False
            break
    if all_consonants_word:
        return word + "ay"
    
    # Check if the word was a title to preserve capitalization after moving the letters
    titlelize = word.istitle()
    word = word.lower()
    
    while word[0] in consonants:
        # Preserve 'qu'
        if word[:2] == "qu":
            word = word[2:] + "qu"
        else:
            word = word[1:] + word[0]
    
    if titlelize:
        word = word.title()
    
    return word + "ay"
```

```python
for word in ("string", "idle", "Cabinet", "quiet"):
    print (word, "→", word_to_pig_latin(word))
```

```python
def text_to_pig_latin(text):
    words = re.findall(r"[a-zA-Z]+", text)
    words = list(set(words))
    words = [w for w in words if len(w) > 2]
    words = sorted(words, key=len, reverse=True)
    for word in words:
        text = text.replace(word, word_to_pig_latin(word))
    return text
```

```python
from nltk.corpus import udhr
text = udhr.raw("English-Latin1")
print(text_to_pig_latin(text))
```
