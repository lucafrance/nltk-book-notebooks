Write a function `novel10(text)` that prints any word that appeared in the last 10% of a text that had not been encountered earlier.

```python
# This is a brute force version, there are definitely optimizations 
def novel10(text):
    n = len(text)
    n_10 = n // 10
    for i in range(n):
        word = text[i]
        if word not in text[:i-n_10] and word not in text[i-n_10:i]:
            print(word)
```

```python
test_text = "a a a a a a a a a a a a b a a a a a a a a a a a b b b b b b b b b b b b c c c c c c c c c c c".split()
novel10(test_text)
```

```python
from nltk.corpus import udhr
text = udhr.words("English-Latin1")
novel10(text)
```
